```python
import pandas as pd
```

```python
df = pd.read_csv('/content/companies.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='state', data=df)
plt.xticks(rotation=60)
plt.show()
```

```python
#if df['job_openings'].isnull().any():
  #  df['job_openings'].fillna(0, inplace=True)
```

```python
df['job_openings'] = pd.to_numeric(df['job_openings'], errors='coerce')
```

```python
industry_jobs = df.groupby('Industry')['job_openings'].sum().reset_index()
```

```python
industry_jobs = industry_jobs.sort_values(by='job_openings', ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Industry', y='job_openings', data=industry_jobs, hue='job_openings', palette='viridis')
plt.title('Total Job Openings by Industry', fontsize=16)
plt.xlabel('Industry', fontsize=14)
plt.ylabel('Total Job Openings', fontsize=14)
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_growth = df.groupby('founded')['growth_percentage'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=avg_growth, x='founded', y='growth_percentage', marker='o', color='blue')
plt.title('Average Growth Percentage by Founded Year', fontsize=16)
plt.xlabel('Founded Year', fontsize=14)
plt.ylabel('Average Growth Percentage', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(visible=True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='employees', y='estimated_revenues', color='green', alpha=0.7)
plt.title('Employees vs Estimated Revenues', fontsize=16)
plt.xlabel('Employees', fontsize=14)
plt.ylabel('Estimated Revenues', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(visible=True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='founded', bins=20, kde=False, color='orange')
plt.title('Number of Companies Founded by Year', fontsize=16)
plt.xlabel('Year Founded', fontsize=14)
plt.ylabel('Number of Companies', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(visible=True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
